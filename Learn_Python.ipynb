```python
from collections import Counter

tags = ["python", "js", "python", "sql", "js", "python"]

counts = Counter(tags)
print(counts)
# Output: Counter({'python': 3, 'js': 2, 'sql': 1})

# Get the top 2 most common items
print(counts.most_common(2))
```

```python
import requests

# Making a GET request to a JSON placeholder API
response = requests.get("https://jsonplaceholder.typicode.com/posts/1")

# Check if the request was successful (Status Code 200)
if response.status_code == 200:
    # Convert the response directly to a Python Dictionary
    post_data = response.json()
    print(f"Title: {post_data['title']}")
else:
    print("Failed to fetch data.")
```

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Hello Full Stack Dev!"}

@app.get("/items/{item_id}")
def read_item(item_id: int, q: str = None):
    return {"item_id": item_id, "query": q}
```
